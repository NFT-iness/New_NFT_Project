### **Abstract**
<hr>

We present the NFT-iness project, an exploration of the activity on the Ethereum non-fungible token (NFT) market from 2017 and early 2023 constructed from the API of Etherscan.

We then consider the following questions:

- Who owns NFTs on the Ethereum blockchain? Are NFTs being traded in a bubble? We explore this idea in the 'commerce of NFT' section of the website.
- What are the trends in NFT market? What is the growth pattern of this market? We explore this idea in the 'trends in NFT' section of the website.

#### RAW DATASETs 
"NFT_RawData_Offset100", "NFT_RawData_Offset1000", "NFT_RawData_Offset10000":

<br>

The columns in the NFT_RawData_Offset are the following:

`Unamed`- a unique ID associated with each event, generated when we created the dataset

`block_number` - the number of the Ethereum transaction block in which the transaction containing the event was mined

`timestamp`- originally a unix timestamp at which the Ethereum transaction block with number block_number was mined into the Ethereum blockchain, which we changed into human readable time format

`hash` - the hash of the Ethereum transaction in which we observed the event

`nonce` - stands for 'number only used once', a sequencially incrementing counter which indicates the transaction number from the account

`blockHash`- hash of the block where the transaction was mined in

`from`- the address that owned the NFT before the Transfer event denoted by this row in the relation (it is not the address that initiated the transaction in which the NFT transfer took place)

`contractAdress` - the address of the smart contract containing the NFT that the event describes

`to`- the address that owned the NFT after the Transfer event denoted by this row in the relation (it is not the address that was the recipient the transaction in which the NFT transfer took place), (null when contract creation transaction)

`token_id `- the identifier that represents the NFT that the event describes within the context of the smart contract with address nft_address

`tokenSymbol`- information what is being send

`tokenDecimal`- number of decimal places used to represent the token's smallest unit

`transactionIndex`- helps identifying events that are logged in one transaction (at this case the transactionIndex has the same value)

`gas` -  fee required to successfully conduct a transaction or execute a contract on the Ethereum blockchain platform

`gasPrice`- Costs for gas

`gasUsed`- Amount of gas used in this transaction

`cumulativeGasUsed`- Cumulated Gas used

`input`- referes to an ERC20token transfer event call

`confirmations`- information about the process of the transaction being included in a block on the blockchain. Showing how many blocks were mined before

`label`- contains one of the following strings: "Collectibles", "Art", "Photography", "DomainNames", "Music", "Virtual Worlds", "Trading Cards", "Utility"

There are over 25000 transactions in the NFT_RawData_Offset10000 Dataset

There are 6,667,282 mints and 4,514,729 transfers in our dataset. Together, these events represent the activity of 9,292 NFT contracts which were actively used on the Ethereum blockchain between April 1, 2021 and September 25, 2021.
